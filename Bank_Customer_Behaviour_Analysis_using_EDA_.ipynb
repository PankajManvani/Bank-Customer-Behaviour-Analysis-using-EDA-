```python
df['Estimated Income'].value_counts()
```

now we will convert this numerical data into categorical data
we will create income band from the estimated income

```python
bins = [0 , 100000 , 300000 , float('inf')]
labels = ['Low', 'Mid', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins,labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#now we'll see the different columns which is also not the numerical columns
#so Examine the distribution of unique categories in categorical columns
categorical_cols = df[["BRId" , "GenderId" , "Amount of Credit Cards" , "IAId" , "Nationality" , "Occupation" , "Fee Structure" , "Loyalty Classification" , "Properties Owned" , "Risk Weighting" , "Income Band"]]

for col in categorical_cols:
  print(f"value counts for '{col}':")
  display(df[col].value_counts())
```

##Univariate analysis

```python
 #this is a simple univariate analysis which means where you are anayzing each column one by one
 #this same peice of code can be used in biariate analysis(here it is biariate analysis cause we are analyzing the relation between two variables)
 #if u remove the hue that means it is purely univariate analysis
for i, predictor in enumerate(categorical_cols):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

##biavariate analysis

```python
#here it means all the factors like number of creditcards etc all are compared with respect to GenderId here 2 means female 1 means male check exacle sheet
for i, predictor in enumerate(categorical_cols):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
#lets do it with respect to the nationality
for i, predictor in enumerate(categorical_cols):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
#HISplot of value counts for different occupation

for col in categorical_cols:
  if col == 'Occupation':
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title("Histogram of Occupation count")
  plt.xlabel(col)
  plt.ylabel("count")
  plt.show()
```

##Numerical Analysis

```python
numerical_cols=['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts',
                'Saving Accounts','Foreign Currency Account','Business Lending']

#univariate analysis and visulization
#plt.figure(figsize=(15,10))
#for i,col in enumerate(numerical_cols):
 #  plt.subplot(2,1,i+1)
  # sns.histplot(df[col],kde=True)
   #plt.title(col)
   #plt.show() video vala code
   #chatgpt vala code:
plt.figure(figsize=(18, 12))  # Width x Height

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)  # 3 rows, 3 columns, current index , places each plot in the right spot of a 3×3 grid.
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()#adjusts spacing to avoid overlap.
plt.show()
```

##Heatmaps

so that we can see the correlation between all the numerical attributes

```python
numerical_cols=['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts',
                'Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

##Insights of EDA:
1.The strongest positive correlation occur among "Bank Deposits" with "checking accounts","Saving accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial  amount/funds accross other accounts as well.

2.Strong Positive Correlations

Checking Accounts & Saving Accounts (0.75): People with high checking account balances also tend to have high savings.

Bank Deposits & Saving Accounts (0.75): Indicates that customers who deposit more money also keep more in savings.

Bank Deposits & Checking Accounts (0.84): Customers with high deposits are likely using both account types actively.

3.Moderate Positive Correlations

Business Lending & Credit Card Balance (0.36) and Business Lending & Bank Loans (0.42): Those using business loans may also rely on credit and personal loans.

Foreign Currency Account has moderate correlations with most financial products (0.30–0.36), suggesting that these accounts are used by well-engaged or wealthier clients.

4.Negative or Weak Correlations

Estimated Income shows weak or negative correlations with most variables, meaning income level doesn’t strongly predict how customers manage their finances in this dataset.

Superannuation Savings and Estimated Income (-0.37): Slight negative correlation, possibly due to people with higher income investing less in this particular savings option (or due to age factors).
